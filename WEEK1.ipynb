### 규칙
- answer 변수에 결과를 저장하세요
- print 사용하지 마세요
- 코드 수정은 문제 셀에서만 하세요

```python
import base64

class HYPER_Grader:
    def __init__(self):
        self.answers = {
            1: "NA==",
            2: "Mw==",
            3: "aGVsbG9fd29ybGQ=",
            4: "UG4=",
            5: "Mw=="
        }

    def check(self, task_id, user_answer):
      try:
          expected = base64.b64decode(self.answers[task_id]).decode()
          result = str(user_answer).strip()
          expected = expected.strip()

          if result == expected:
              print(f"Task {task_id}: ✅ 정답입니다!")
          else:
              print(f"Task {task_id}: ❌ 틀렸습니다.")
              print("당신의 출력:", result)
              print("정답:", expected)

      except Exception as e:
        print(f"Task {task_id}: ⚠ 에러 발생 ({e})")

grader = HYPER_Grader()
print("채점 준비 완료!")
```

## Task 1

문자열 `"hi"` 를 **2번 반복한 문자열의 길이**를 구하세요.

```python
# 여기에 코드를 작성하세요

s = "hi"

answer = len(s*2)
```

```python
grader.check(1, answer)
```

## Task 2

문자열 `"banana"` 에서 **'a'의 개수**를 구하세요.

```python
s = "banana"

answer = s.count('a')
```

```python
grader.check(2, answer)
```

## Task 3

```python
a = "Hello"
b = "World"
a=a+"_"
print(a)
answer = a.replace("H","h") + b.replace("W","w")
print(answer)
```

```python
grader.check(3, answer)
```
